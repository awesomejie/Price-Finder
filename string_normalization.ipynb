#### What to do with '-' in texts?
1. replace '-' with space  
   Ex. 3.0-inch -> 3.0 inch  
   Nikon EN-EL9a 1080mAh -> Nikon EN EL9a 1080mAh
   
2. leave '-' untouched  
   Bad similarity score between "canon li-ion", "canon liion", but bad effect reduces for longer strings, e.g.,  
   "nikon en-el5 1100mah ultra high capacity li-ion battery pack for nikon p100 p90 p80 p6000"  
   vs  
   "nikon en-el5 1100mah ultra high capacity liion battery pack for nikon p100 p90 p80 p6000"
   
3. remove '-'  
   Bad similarity score between "canon lithium ion", "canon lithiumion"

```python
exclude = set(string.punctuation)
include = set('.-')
exclude -= include

cachedStopWords = set(stopwords.words("english"))

def handle_strings(x):
    """
    Helper function to make string all lowercase and remove punctuation & stopwords.
    
    x: any string
    """
    x = x.strip()
    x = x.lower()
    x = x.replace('/', ' ')  # replace '/' with space
    #x = x.replace('-', ' ')  # replace '-' with space
    x = ''.join(ch for ch in x if ch not in exclude)
    x = ' '.join(word for word in x.split() if word not in cachedStopWords)
    return x

for i in range(50):
    print(listings[i]['title'])
    print(handle_strings(listings[i]['title']))
    print("=")
```
